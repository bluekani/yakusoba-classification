```
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from keras.preprocessing.image import img_to_array, load_img
import os
```

```
input_model = 'drive/My Drive/Colab Notebooks/model_ac.h5f'
image_dir = 'drive/My Drive/Colab Notebooks/image/picture'
```

```
model=load_model(input_model)
input_size = 64
label=['Gotsu_shio','Gotsu_source','Bubuka','Ippei','U.F.O.']

for file in os.listdir(image_dir):
    image_list = []
    filepath = image_dir + "/" + file

    img = img_to_array(load_img(filepath, target_size=(input_size,input_size)))
    plt.figure()
    plt.imshow(img/255)

    image_list.append(img)
    image_list = np.array(image_list)

    pred = model.predict(image_list, batch_size=1, verbose=0)
    score = np.max(pred)
    pred_label = label[np.argmax(pred[0])]
    print('name:',pred_label)
    print('score:',score)
    print()
```
